```python
# Function to remove nouns from a sentence
def remove_nouns(sentence):
    # Tokenize the sentence
    words = word_tokenize(sentence)
    
    # POS tagging
    pos_tags = pos_tag(words)
    
    # Filter out nouns (NN: singular noun, NNS: plural noun, NNP: proper noun, NNPS: proper plural noun)
    words_without_nouns = [word for word, pos in pos_tags if pos not in ['NN', 'NNS']]
    
    # Join the remaining words back into a sentence
    result_sentence = ' '.join(words_without_nouns)
    
    return result_sentence
```
